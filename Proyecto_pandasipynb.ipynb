```python
datos.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor', ascending=False)
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor', ascending=False)
df_tipo_precio.plot(kind='bar',figsize=(12,8),color='red')
```
